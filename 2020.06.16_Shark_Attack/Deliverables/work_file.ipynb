```python
# simplify column names: remove dots, spaces at both ends, lower case

def clean_column(df_raw):
        column_clean = []
        for column in df_raw.columns:
            column = column.replace('.',' ')
            column = column.strip().lower()
            column_clean.append(column)
        
        df_raw.columns = column_clean
        return df_raw.columns
```
